# Dopplereffekt for lyd og lys

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Dopplereffekt for lyd

```python
x = (1/(1-(-20/340)))*800
print(x)
```

```python
# Definition af variable
t = np.linspace(0, 1, 1000)
v_lyd = 343
v_sender = 10
f = 1000

# Oprindlig vinkelfrekvens
omega_n = 2*np.pi*f

# Dopplerforskudt vinkelfrekvens
f1 = (1/(1-(v_sender/v_lyd)))*f
omega_d = 2*np.pi*f1

print(omega_d)

# Nye y-værdier
n = np.sin(omega_n*t)
d = np.sin(omega_d*t)

# Plotting
fig, ax = plt.subplot_mosaic([["oprindlig"],["doppler"],["samlet"]])

ax["oprindlig"].plot(t,n, label="original bølge")
ax["doppler"].plot(t,d, label="doppler-forskudt bølge")

ax["samlet"].plot(t,n, label="original bølge")
ax["samlet"].plot(t,d, label="doppler-forskudt bølge")

ax["samlet"].legend(loc="upper right")
ax["samlet"].set_xlabel("tid [s]")

ax["oprindlig"].grid()
ax["doppler"].grid()
ax["samlet"].grid()
plt.show()
```

## Dopplereffekt for lys

```python
# Definition af variable
t = np.linspace(0, 1, 1000)
v = 1000000
c = 3 * 10**8
f = 1000

# Oprindlige vinkelfrekvens 
omega_n = 2*np.pi*f

# Dopplerforskudt vinkelfrekvens
f1 = np.sqrt((1+(v/c))/(1-v/c))*f
omega_d = 2*np.pi*f1

# Udregning at y-værdierne
n = np.sin(omega_n*t)
d = np.sin(omega_d*t)

# Plotting
fig, ax = plt.subplot_mosaic([["oprindlig"],["doppler"],["samlet"]])
ax["oprindlig"].plot(t,n, label="original bølge")
ax["doppler"].plot(t,d, label="doppler-forskudt bølge")

ax["samlet"].plot(t,n, label="original bølge")
ax["samlet"].plot(t,d, label="doppler-forskudt bølge")

ax["samlet"].legend(loc="upper right")
ax["samlet"].set_xlabel("tid [s]")

ax["oprindlig"].grid()
ax["doppler"].grid()
ax["samlet"].grid()
plt.show()
```
